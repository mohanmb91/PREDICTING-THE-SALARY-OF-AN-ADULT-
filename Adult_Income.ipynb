# 2 pts Create a notebook, import Pandas, and import the data as a dataframe.

==> In next two cells I have Imported all the required package and also using pandas CSV function I have imported the dataset as well.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('C:\\Users\\Mohan Kumar\\Downloads\\data\\adult.csv')
data
```

# 2 pts Copy all columns with empty values and replace all empty numeric fields with the average of its column in this new column.

==> copied all the numeric feilds to new numeric fields and replaced the zero's with the mode(it takes the most repeated value) of the column.

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
newdf = data.select_dtypes(include=numerics)
column_list = newdf.columns.values
for each_column in column_list:
    data[str(each_column) + '_new'] = newdf[str(each_column)].replace(0,newdf[str(each_column)].mode())
data
```

# 2pts Convert non-numeric columns to columns with an integer representations.

==> using FIT TRANSFORM of preprocessing label encoder.
==> And also used used Describe to check the row count and other details.

```python
le = preprocessing.LabelEncoder()
newdf1 = data.select_dtypes(['object'])
for each_non_numeric in newdf1.columns.values:
    data[each_non_numeric] = le.fit_transform(data[each_non_numeric].values)
data
```

```python
data.describe()
```

# 2 pts Convert the data to a numpy array

==> using Numpy converted the data into array's.

==> training data contains all the array of the given datasets which exclude's INCOME column(Vector result)

```python

training_data = np.array(data)
training_data = training_data[:,[3,5,6,7,8,9,13,15,16,17,18,19,20]]
training_data
```

==> In training Result I am storing the INCOME(vector result)

```python
training_result = np.array(data)
training_result = training_result[:,[14]]
training_result
```

# 2 pts Divide the data into a training and test set.

==>here we are splitting the data into training and test set.

```python
from sklearn.cross_validation import train_test_split
training_data1, test_data, training_result1, test_result = train_test_split(training_data, training_result, test_size=0.33, random_state=0)
training_data1
```

```python
training_result1
```

# 2 pts Run a random forest on the training set.

==> here Random forest has been trainned with training data and training result using FIT method

```python
model = RandomForestClassifier(n_estimators=100 , oob_score = True , random_state = 42)
model.fit(training_data1,training_result1.ravel())
```

# 2 pts Compute the accuracy of predictions on the test set.

==> here we are using the model which we have created with training data and we are using the test data  to predict the vector result.

==>since, we have generated the output which was predicted by Random Forest we need to compare the generated output with test _ result for accuracy score(this is done using metrics accuracy). 

```python
output = model.predict(test_data)
output
```

```python
metrics.accuracy_score(output,test_result)
```

# classification Report on Income(> 50 and < 50).

```python
from sklearn.metrics import classification_report
target_names = ['<50','>50']
classification_report(test_result,output,target_names=target_names)
```
